### RFM Metrics

```python
df["last_order_date"].max()
```

```python
today_date = dt.datetime(2021, 6, 1)
today_date
```

```python
rfm = df.groupby("master_id").agg({"last_order_date": lambda x: (today_date - x).astype('timedelta64[D]'),
                                   "order_num_total": "sum",
                                   "customer_value_total": "sum"})
rfm.head()
```

```python
rfm.columns = ["recency","frequency","monetary"]
rfm = rfm.reset_index()
rfm.rename(columns = {'master_id':'customer_id'}, inplace = True)
rfm.head()
```

### RFM Scores

```python
rfm["recency"] = pd.qcut(rfm["recency"],5,[5,4,3,2,1])
rfm["frequency"] = pd.qcut(rfm["frequency"].rank(method= "first"),5,[1,2,3,4,5])
rfm["monetary"] = pd.qcut(rfm["monetary"],5,[1,2,3,4,5])
rfm.head()
```

```python
rfm["RF_SCORE"] = (rfm["recency"].astype(str) + rfm["frequency"].astype(str))
rfm.head()
```

### RFM Labeling With REGEX

```python
seg_map = {
        r'[1-2][1-2]': 'hibernating', #1. elemanında 1yada2 2.elemanında 1yada2 görürsen hibernating yaz
        r'[1-2][3-4]': 'at_risk', #1.elemanında 1yada2 2.elemanında 3yada4 görürsen at risk yaz
        r'[1-2]5': 'cant_loose', #1.elemanında 1yada2 2.elemanında 5 görürsen cant_loose
        r'3[1-2]': 'about_to_sleep',
        r'33': 'need_attention', #1.elemanında 3 2.elemanında 3 görürsen need_attention yaz
        r'[3-4][4-5]': 'loyal_customers',
        r'41': 'promising',
        r'51': 'new_customers',
        r'[4-5][2-3]': 'potential_loyalists',
        r'5[4-5]': 'champions'
    }
rfm["segment"]= rfm["RF_SCORE"].replace(seg_map, regex = True)
rfm.head()
```

```python
# FLO bünyesine yeni bir kadın ayakkabı markası dahil ediyor. Dahil ettiği markanın ürün fiyatları genel müşteri
# tercihlerinin üstünde. Bu nedenle markanın tanıtımı ve ürün satışları için ilgilenecek profildeki müşterilerle özel
# olarak iletişime geçmek isteniliyor. Sadık müşterilerinden(champions, loyal_customers) ve kadın kategorisinden
# alışveriş yapan kişiler özel olarak iletişim kurulacak müşteriler. Bu müşterilerin id numaralarını csv dosyasına
# kaydediniz.
```

```python
champ_loyal = rfm[(rfm["segment"] == "champions")|(rfm["segment"] == "loyal_cutomers")]["customer_id"]
```

```python
cust_ids = df[(df["master_id"].isin(champ_loyal))&(df["interested_in_categories_12"].str.contains("KADIN"))]["master_id"]
cust_ids.to_csv("yeni_marka_hedef_müşteri_id.csv", index=False)
cust_ids.shape
```
